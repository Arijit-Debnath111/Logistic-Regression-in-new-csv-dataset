```python
data['y'] = np.where(data['y'] == 'yes',1,0)
data['y'] = data['y'].astype(str)
```

```python
data.head(10)
```

```python
col_list = list(data.columns)
```

```python
col_list.remove('y')
```

```python
for col in col_list:
    if data[col].dtypes=='object':
        #print('ob')
        data[col] = data[col].fillna(data[col].mode()[0])
    else:
        #print('num')
        data[col] = data[col].fillna(data[col].mean())
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
data = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
data.head(10)
```

```python
data.shape
```

```python
##Binning of Age##
bins = [0, 1, 5, 10, 25, 50, 100]
data['age'] = pd.cut(data['age'], bins)

data['age'] = data.age.astype(str)
```

```python
data['age'].unique()
```

```python
data.head()
```

```python
col_list = []
for col in data.columns:
    if ((data[col].dtype == 'object') & (col != 'y') ):
        col_list.append(col)

df_2 = pd.get_dummies(data[col_list],drop_first=True)

for col in df_2.columns:
    df_2[col] =  df_2[col].astype(int)
df_2.shape
```

```python
df_2.head()
```

```python
df_combined = pd.concat([data, df_2], axis=1)
df_combined.shape
```

```python
df_combined.head()
```

```python
col_list
```

```python
###Drop fields for which dummy vars already created
df_combined.drop(columns = col_list,axis = 1,inplace = True)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor 
col_list = []
for col in df_combined.columns:
    if ((df_combined[col].dtype != 'object') & (col != 'y') ):
        col_list.append(col)

X = df_combined[col_list]
vif_data = pd.DataFrame() 
vif_data["feature"] = X.columns 
vif_data["VIF"] = [variance_inflation_factor(X.values, i) 
                          for i in range(len(X.columns))]  
print(vif_data)
```
